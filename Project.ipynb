# **Phân tích Chiến tranh Nga-Ukraina [2022-2024]**

Chiến tranh Nga-Ukraine là cuộc xung đột đang diễn ra giữa Nga (cùng với lực lượng ly khai thân Nga) và Ukraine. Nó bắt đầu vào tháng 2 năm 2014 sau Cách mạng Nhân phẩm ở Ukraine, ban đầu tập trung vào tình trạng của Crimea và Donbas, được quốc tế công nhận là một phần của Ukraine. Tám năm đầu tiên của cuộc xung đột bao gồm việc Nga sáp nhập Crimea (2014) và cuộc chiến ở Donbas (2014-2022) giữa Ukraine và những người ly khai được Nga hậu thuẫn, cũng như các sự cố trên biển, tấn công mạng và căng thẳng chính trị. Kể từ cuối năm 2021, với việc Nga tăng cường quân đội ở biên giới Ukraine, cuộc xung đột đã leo thang đáng kể khi Nga tiến hành một cuộc xâm lược toàn diện vào ngày 24 tháng 2 năm 2022. Cuộc xâm lược này đã gây ra cuộc khủng hoảng người tị nạn lớn nhất ở châu Âu kể từ Thế chiến II, với hơn 8,8 triệu người Ukraine chạy trốn khỏi đất nước và một phần ba dân số phải di dời. Tại thời điểm tiến hành phân tích này, khoảng 18% lãnh thổ Ukraine vẫn nằm dưới sự chiếm đóng của Nga. Điều này bao gồm Bán đảo Crimea cũng như phần lớn các khu vực Donetsk và Luhansk. Bất chấp những nỗ lực đáng kể của Ukraine và sự hỗ trợ của các đồng minh, tiền tuyến, trải dài gần 1.000 km, không thay đổi nhiều trong những tháng gần đây, cho thấy cuộc xung đột đang "sa lầy".

```python
# Tạo khung dữ liệu mới
directions = pd.DataFrame({
    'places': [
        'bakhmut', 'donetsk', 'kramatorsk', 'avdiivka', 'izyum', 'lyman',
        'kryvyi rih', 'kurakhove', 'zaporizhzhia', 'sloviansk',
        'kharkiv', 'sievierodonetsk', 'mykolaiv', 'popasna', 'novopavlivsk'
    ],
    'frequency': [0] * 15
})

directions
```

```python
def update_counts(df1, directions):
    """Updates the occurrence count of specific keywords in df1 and records them in the places DataFrame under the 'frequency' column."""
    keywords = directions['places'].tolist()
    for keyword in keywords:
        frequency = df1['greatest losses direction'].str.contains(keyword, case=False, na=False).sum()
        directions.loc[directions['places'] == keyword, 'frequency'] += frequency

update_counts(df1, directions)
directions
```

```python
# Lấy hàng đầu tiên từ cột thứ ba trở đi.
loc1 = df_combo.iloc[:, 2:]

# Chuyển đổi tất cả các cột thành dạng số, các lỗi sẽ được chuyển thành NaN.
c1 = loc1.apply(pd.to_numeric, errors='coerce')

# Lấy giá trị lớn nhất của mỗi cột.
cumulative_losses1 = c1.max()

# Chuyển đổi sang DataFrame và sắp xếp.
cumulative_losses1 = cumulative_losses1.reset_index()

# Đổi tên các cột.
cumulative_losses1.columns = ['unit', 'unit_amount']

# Sắp xếp các cột theo giá trị 'unit_amount'.
cumulative_losses1 = cumulative_losses1.sort_values(by='unit_amount', ascending=False)
cumulative_losses1
```

```python
air_units = ['aircraft', 'helicopter', 'drone', 'cruise missiles']
naval_units = ['naval ship', 'submarines']
land_units = ['tank', 'APC', 'field artillery', 'MRL', 'anti-aircraft warfare', 'special equipment',
               'greatest losses direction', 'vehicles and fuel tanks']
personnel_units = ['killed in action', 'prisoners']
zones = ['greatest losses direction']

# Tính tổng số đơn vị cho mỗi danh mục và thêm các cột mới.
df_combo['total air units'] = df_combo[air_units].sum(axis=1, numeric_only=True)
df_combo['total ground units'] = df_combo[land_units].sum(axis=1, numeric_only=True)
df_combo['total naval units'] = df_combo[naval_units].sum(axis=1, numeric_only=True)
df_combo['zones'] = df_combo[zones].sum(axis=1, numeric_only=True)
df_combo['total personnel units'] = df_combo[personnel_units].sum(axis=1, numeric_only=True)
```

```python
def impute_unit(unit_name):
    """
    Classifies unit names into different categories based on predefined lists (air_units, naval_units, personnel_units, zones),
    and then applies this function to the 'unit' column of the DataFrame cumulative_losses1 using the apply method along with a lambda function to create a new column 'unit_type'.
    This new column contains the assigned unit category for each row based on the unit name.
    """
    if unit_name in air_units:
        return 'Air Units'
    elif unit_name in naval_units:
        return 'Naval Units'
    elif unit_name in personnel_units:
        return 'Personnel Units'
    elif unit_name in zones:
        return 'Zones'
    else:
        return 'Land Units'


cumulative_losses1['unit_type'] = cumulative_losses1['unit'].apply(lambda x: impute_unit(x))
cumulative_losses1
```

```python
# Lọc dữ liệu
filtered_cumm_sum_df = cumulative_losses1[cumulative_losses1['unit_type'] != 'Zones']

# Nhóm theo 'unit_type' và tính tổng 'unit_amount'.
unit_type_sums = filtered_cumm_sum_df.groupby('unit_type')['unit_amount'].sum().reset_index()


plt.figure(figsize=(8, 8))

# Tạo biểu đồ hình tròn bằng matplotlib
sizes = unit_type_sums['unit_amount']
labels = unit_type_sums['unit_type']
colors = sns.color_palette("RdBu", len(labels))

# Tạo một hình nêm cho mỗi nhãn, hiển thị phần trăm bên trong biểu đồ
plt.pie(sizes, labels=None, colors=colors, startangle=90, wedgeprops=dict(width=0.4), autopct='%1.3f%%', pctdistance=0.85)

plt.title('Total Losses Distribution', fontsize=16, fontweight='bold')


plt.gcf().set_facecolor((1, 230/255, 188/255, 1))
plt.gca().set_facecolor('white')

plt.legend(title='Unit Type', labels=labels, loc='upper left', fontsize=12, title_fontsize=14, bbox_to_anchor=(1.05, 1), borderaxespad=0.)

plt.show()
```

Việc mất mát nhiều tên lửa hành trình cũng có thể ảnh hưởng đến nguồn lực dự trữ của Nga, đặc biệt khi việc sản xuất và bổ sung loại vũ khí phức tạp này không dễ dàng và có thể chịu tác động từ các lệnh cấm vận quốc tế.
